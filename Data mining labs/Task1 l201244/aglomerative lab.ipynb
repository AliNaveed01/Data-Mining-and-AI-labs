# Agglomerative clustering 

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

from collections import Counter
```

```python
# Perform K-medoids clustering on basic2.csv
file1 = pd.read_csv('basic2.csv')
file2 = pd.read_csv('blob.csv')
file3 = pd.read_csv('boxes3.csv')
file4 = pd.read_csv('dart.csv')
file5 = pd.read_csv('outliers.csv')
file6 = pd.read_csv('spiral2.csv')
```

```python
# now we will implement the agglomerative clustering algorithm using the euclidean distance
# we will use the single linkage method

from scipy.spatial.distance import pdist, squareform
from sklearn.cluster import AgglomerativeClustering

# for dataset 1

ag = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
ag.fit(file1)

#visualizing the clusters
plt.scatter(file1['x'], file1['y'], c=ag.labels_, cmap='rainbow')
plt.show()

```

```python
ag = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='average')
ag.fit(file2)

#visualizing the clusters
plt.scatter(file2['x'], file2['y'], c=ag.labels_, cmap='rainbow')
plt.show()
```

```python
ag = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
ag.fit(file4)

#visualizing the clusters
plt.scatter(file4['x'], file4['y'], c=ag.labels_, cmap='rainbow')
plt.show()
```

```python
ag = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
ag.fit(file3)

#visualizing the clusters
plt.scatter(file3['x'], file3['y'], c=ag.labels_, cmap='rainbow')
plt.show()
```
